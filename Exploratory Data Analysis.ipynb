# Top 50 Oldest Players

```python
Top_10_oldest_players = data[['Name','Age','Nationality']].sort_values(by='Age',ascending=False)[:50]
```

```python
plt.figure(figsize=(12,3))
#plt.bar(x=Top_10_oldest_players['Name'],height=Top_10_oldest_players['Age'])
#plt.xticks(rotation=90)
Top_10_oldest_players[['Name','Age']].plot(kind='barh',figsize=(15,9))
plt.yticks(range(0,len(Top_10_oldest_players['Name'])),Top_10_oldest_players['Name'])
plt.xlabel('Players Name')
plt.ylabel('Age')
plt.title("Ten 50 oldest players age",fontsize=20)
plt.show()
```

# Overall best players

```python
overall_best = data[['Name','Club','Overall']].sort_values(by='Overall',ascending=False)[:10]
```

```python
plt.figure(figsize=(12,10))
plt.pie(x=overall_best['Overall'],labels=overall_best['Name'],autopct='%1.2f%%')
plt.title('Top 10 Overall Rank Footballers',fontsize=20)
plt.show()
```

# Top 10  Nationailty with valued players


```python
Top_10_Nationailty_with_valued_players = data.groupby('Nationality', as_index=False)['Value'].sum()[:10]
```

```python
Top_10_Nationailty_with_valued_players = Top_10_Nationailty_with_valued_players.sort_values(by='Value',ascending=False)
```

```python
Top_10_Nationailty_with_valued_players = Top_10_Nationailty_with_valued_players.reset_index(drop=True)
```

```python
plt.figure(figsize=(20,15))
plt.pie(x=Top_10_Nationailty_with_valued_players['Value'],labels=Top_10_Nationailty_with_valued_players['Nationality'],autopct='%1.2f%%')
plt.title('Top 10 Nationality with Valued Players',fontsize=25)
plt.show()
```
